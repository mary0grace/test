```python
import numpy as np 
import matplotlib.pyplot as plt
import sympy as sp

%matplotlib inline 
```

```python
θ = sp.Symbol('θ')
T = sp.Symbol ('T')

ε_AA = 238 # кал/моль
σ_AA = 3.4 # Ангстрем - диаметр жёсткой сферы молекулы
r = 2**(1/6)*σ_AA
ε = 4*ε_AA*((σ_AA/r)**6-(σ_AA/r)**12)
R = 2 # кал/(моль*К) - газовая постоянная 
T_kr = 146.6 # К - критическая температура
#T = 146.6 # начальное значение температуры
z_1 = 6
z_2 = 12

f1 = θ*(1+2*θ*(sp.exp(-(R*T)**(-1)*ε)-1)/(1+(sp.exp(-(R*T)**(-1)*ε)-1)*(1-2*θ)+((1+(sp.exp(-(R*T)**(-1)*ε)-1)*(1-2*θ))**2+4*(sp.exp(-(R*T)**(-1)*ε)-1)*θ**2)**0.5))**z_1/(1-θ)
f1 = np.log(θ*(1+2*θ*(sp.exp(-(R*T)**(-1)*ε)-1)/(1+(sp.exp(-(R*T)**(-1)*ε)-1)*(1-2*θ)+((1+(sp.exp(-(R*T)**(-1)*ε)-1)*(1-2*θ))**2+4*(sp.exp(-(R*T)**(-1)*ε)-1)*θ**2)**0.5))**z_1/(1-θ))

f = sp.lambdify([θ, T], f1)

#f1_sym=f1(sp.Symbol('θ'),sp.Symbol('T'))
#f1_diff1_θ = f1.subs([(T, 100),(θ, 0.5)])
```
